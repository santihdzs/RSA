```python

import matplotlib.pyplot as plt
import numpy as np


# Perform linear regression
coefficients = np.polyfit(firstPrimeValues, character1Values, 1)
poly = np.poly1d(coefficients)

# Calculate Pearson correlation coefficient (r)
correlation_matrix = np.corrcoef(firstPrimeValues, character1Values)
r_squared = correlation_matrix[0, 1] ** 2

# Set up the figure and axes
fig, ax = plt.subplots()

# Plot the scatter plot
ax.scatter(firstPrimeValues, character1Values, marker='o', color='b', label='Data Points')

# Plot the line of best fit
ax.plot(firstPrimeValues, poly(firstPrimeValues), color='r', label=f'Line of Best Fit: {coefficients[0]:.2f}x + {coefficients[1]:.2f}')

# Set x-axis range and markers
ax.set_xticks(range(0, 8001, 1000))
ax.set_xlim(0, 8000)

# Set y-axis range and markers
ax.set_yticks([i * 1e7 for i in range(11)])
ax.set_ylim(0, 1e8)

# Set labels and title
ax.set_xlabel('First Prime Values')
ax.set_ylabel('Character 1 Values')
# Set title with properly formatted LaTeX string
title = r'Scatter Plot with Line of Best Fit\nPearson Correlation Coefficient ($r^2$): {:.4f}'.format(r_squared)
title = title.replace(r'\n', r'\\')

# Set title with properly formatted LaTeX string
ax.set_title(title)


# Show grid for better readability
ax.grid(True)

# Add legend
ax.legend()

# Show the plot
plt.show()

# Return the Pearson correlation coefficient (r^2)
r_squared
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the product of consecutive pairs in firstPrimeValues - 1
product_values = [(n1 * n2 - 1) for n1, n2 in zip(firstPrimeValues[:-1], firstPrimeValues[1:])]

# Set up the figure and axes
fig, ax = plt.subplots()

# Plot the values using firstPrimeValues as x and product_values as y
ax.plot(firstPrimeValues[:-1], product_values, label=r'Maximum Encrypted Output ($P_n \cdot P_{n+1} - 1$)', color="k")

# Shade the area above the line in the same color
ax.fill_between(firstPrimeValues[:-1], product_values, max(product_values) * 1.1, color="red", alpha=0.2)

# Shade the area below the line in a different color
ax.fill_between(firstPrimeValues[:-1], product_values, 0, color="green", alpha=0.2)

existing_ticks = [i * 1000 for i in range(8)]  # Your existing ticks

# Add two specific ticks (e.g., 3 and 7908)
additional_ticks = [3, 7907]

# Combine existing ticks with additional ticks
combined_ticks = existing_ticks + additional_ticks

# Set x-axis ticks
ax.set_xticks(combined_ticks)
ax.set_xlim(3, 7907)

# Set y-axis range and markers
num_yticks = 10
ytick_values = np.linspace(0, max(product_values) * 1.1, num_yticks)
ax.set_yticks(ytick_values)
ax.set_ylim(0, max(product_values)*1.1)

# Set labels and title
ax.set_xlabel(r'Prime Numbers ($P_n$)')
ax.set_ylabel('Encrypted Output')
ax.set_title('Prime Numbers vs Encrypted Output')

# Add legend
ax.legend()


ax.text(5500, 17000000, 'Possible Output')
ax.text(2000, 40000000, 'Impossible Output')


# Show the plot
plt.savefig('../figures/comparison/prime_comparison.png')
plt.show()
```
